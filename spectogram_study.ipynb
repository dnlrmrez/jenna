```python
import numpy as np
import matplotlib.pyplot as plt
import os
import librosa
import librosa.display
import IPython.display as idp
```

```python
jenna_path = "C:/Users/dnlrm/tmp/jenna.wav"
idp.Audio(jenna_path)
```

```python
jenna_sig, sr = librosa.load(jenna_path, sr=44100)
jenna_sig.shape
```

```python
sr
```

```python
FRAME_SIZE = 2048
HOP_SIZE = 512
s_jenna = librosa.stft(jenna_sig, n_fft=FRAME_SIZE, hop_length=HOP_SIZE)
s_jenna.shape
```

```python
type(s_jenna[0][0])
```

```python
Y_scale = np.abs(s_jenna) ** 2
Y_scale.shape
```

```python
type(Y_scale[0][0])
```

```python
def plot_spectogram(Y, sr, hop_length, y_axis="linear"):
    plt.figure(figsize=(25,10))
    librosa.display.specshow(Y,
                             sr=sr,
                             hop_length=hop_length,
                             x_axis="time",
                             y_axis=y_axis)
    plt.colorbar(format="%+2.f")
```

```python
plot_spectogram(Y_scale, sr, HOP_SIZE)
```

```python
Y_log_scale = librosa.power_to_db(Y_scale)
plot_spectogram(Y_log_scale, sr, HOP_SIZE)
```

```python
Y_log_scale.shape
```

```python
plot_spectogram(Y_log_scale, sr, HOP_SIZE, y_axis="log")
```
